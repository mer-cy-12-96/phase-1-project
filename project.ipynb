```python
# importing the necessary library

import pandas as pd
```

```python
# Loading the datasets

file_path1 = r"C:\Users\ADMIN\Downloads\archive\AviationData.csv"
file_path2 = r"C:\Users\ADMIN\Downloads\archive\USState_Codes.csv"

df1 = pd.read_csv(file_path1, encoding='ISO-8859-1')
df2 = pd.read_csv(file_path2, encoding='ISO-8859-1')

print(df1.head())
print(df2.head())
```

```python
print("First CSV File - AviationData:")
print(df1.head())

print("\nSecond CSV File - USState_Codes:")
print(df2.head())
```

```python
print("Missing values in AviationData:\n", df1.isnull().sum())
print("\nMissing values in USState_Code:\n", df2.isnull().sum())
```

```python
print("AviationData Info:")
print(df1.info())

print("\nUSState_Codes Info:")
print(df2.info())
```

```python
print("Duplicate rows in AviationData:", df1.duplicated().sum())
print("Duplicate rows in USState_Codes:", df2.duplicated().sum())
```

```python
print("Columns in AviationData:\n", df1.columns)
print("\nColumns in USState_Codes:\n", df2.columns)
```

```python
merged_df = pd.merge(df1, df2, left_on="Location", right_on="US_State", how="left")

print(merged_df.head())
```

```python
print("Missing values in merged dataset:\n", merged_df.isnull().sum())
```

```python
merged_df["Location"].fillna("Unknown", inplace=True)
merged_df["Country"].fillna("Unknown", inplace=True)
merged_df["US_State"].fillna("Non-US", inplace=True)
merged_df["Abbreviation"].fillna("Non-US", inplace=True)
```

```python
injury_columns = ["Total.Fatal.Injuries", "Total.Serious.Injuries", "Total.Minor.Injuries", "Total.Uninjured"]
merged_df[injury_columns] = merged_df[injury_columns].fillna(0)
```

```python
merged_df["Injury.Severity"].fillna("Unknown", inplace=True)
```

```python
merged_df["Aircraft.damage"].fillna("Unknown", inplace=True)
```

```python
columns_to_fill = ["Aircraft.Category", "Engine.Type", "FAR.Description", "Schedule", "Purpose.of.flight"]
merged_df[columns_to_fill] = merged_df[columns_to_fill].fillna("Unknown")
```

```python
print("Missing values after cleaning:\n", merged_df.isnull().sum())
```

```python
print(merged_df.info())
print(merged_df.describe())
print(merged_df.head())
```

```python
print(merged_df["Broad.phase.of.flight"].value_counts().head(10))
```

```python
import matplotlib.pyplot as plt

# count the number of accidents per country
top_countries = merged_df["Country"].value_counts().head(10)

# create a bar chart
plt.figure(figsize=(10, 5))
plt.bar(top_countries.index, top_countries.values, color='skyblue')
plt.title("Top 10 Countries with Most Accidents")
plt.xlabel("Country")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# convert 'Event.Date' to datetime format
merged_df["Event.Date"] = pd.to_datetime(merged_df["Event.Date"], errors="coerce")

# Extract year from the date
merged_df["Year"] = merged_df["Event.Date"].dt.year

# check if it worked
print(merged_df["Year"].head())
```

```python
import matplotlib.pyplot as plt

# count the number of accidents per year
accidents_per_year = merged_df["Year"].value_counts().sort_index()

#plot the trend
plt.figure(figsize=(10, 5))
plt.plot(accidents_per_year.index, accidents_per_year.values, marker='o', linestyle='-')

# labels and title
plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.title("Avition Accidents Over the Years")
plt.grid(True)

# show the plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# count accidents by phase of flight
merged_df["Broad.phase.of.flight"].value_counts().plot(kind="bar", color="skyblue")

# labels and title
plt.xlabel("Phase of Flight")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Phase of Flight")
plt.xticks(rotation=45)

# show plot
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# count accidents by injury severity
merged_df["Injury.Severity"].value_counts().plot(kind="bar", color="pink")

# labels and title
plt.xlabel("Injury Severity")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Injury Severity")
plt.xticks(rotation=45)

# show plot
plt.show()
```
